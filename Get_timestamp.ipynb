```python
def get_timestamp_for_the_post(link, brand_df):
    filtered_df = brand_df[brand_df['url'] == link]

    if filtered_df.empty:
        return None, None, None
    else:
        no_of_likes = filtered_df['likesCount']
        no_of_likes = no_of_likes.iloc[0]
        print("Mention_type", type(filtered_df['mentions']))
        print("Mention", filtered_df['mentions'])
        follower_count = get_follower_count(filtered_df['mentions'])
        print("follower", follower_count)
        timestamp = filtered_df.iloc[0]['timestamp']
        print(timestamp)
        if follower_count is None:
            return timestamp, no_of_likes, None
        return timestamp, no_of_likes, follower_count
```

```python
def get_total_commment_counts(timestamp):
    one_week_after = timestamp + pd.to_timedelta(7, unit='D')
    filtered_comments = comments_df[(comments_df['timestamp'] >= timestamp) & (comments_df['timestamp'] <= one_week_after)]
    no_of_comments = filtered_comments.shape[0]
    return no_of_comments
```

```python
def engagement_ratio(likes, no_of_comments, followers):
    return (likes + no_of_comments) / followers
```

```python
all_engagement = []

#converting the timestamp field of brand and comments df 
brand_df['timestamp'] = pd.to_datetime(brand_df['timestamp'], utc=True)
comments_df['timestamp'] = pd.to_datetime(comments_df['timestamp'], utc=True)

for i in range(len(comments_df)):
    link = comments_df['postUrl'].iloc[i]
    # inf_engagement = []
    # for _ in range(len(brand_df)):
    timestamp, no_of_likes, no_of_followers = get_timestamp_for_the_post(link, brand_df)
    if timestamp is None and no_of_likes is None and no_of_followers is None:
        print("None")
        continue
    else:
        no_of_comments = get_total_commment_counts(timestamp)
        engagement = engagement_ratio(no_of_likes, no_of_comments, no_of_followers)
        print("engagement = ", engagement)
        # inf_engagement.append(engagement)
        all_engagement.append(engagement)
    
```
