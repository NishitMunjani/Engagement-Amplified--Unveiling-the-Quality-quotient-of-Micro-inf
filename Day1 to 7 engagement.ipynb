```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with 7 columns

# Sample DataFrame creation
data = {
    'Column1': [1, 2, 3, 4, 5],
    'Column2': [2, 3, 4, 5, 6],
    'Column3': [3, 4, 5, 6, 7],
    'Column4': [4, 5, 6, 7, 8],
    'Column5': [5, 6, 7, 8, 9],
    'Column6': [6, 7, 8, 9, 10],
    'Column7': [7, 8, 9, 10, 11]
}

df = pd.DataFrame(data)

# Transpose the DataFrame
df_transposed = df.T

# Plotting each column independently
for col in df_transposed.columns:
    plt.scatter(df_transposed.index, df_transposed[col], label=col)

plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.title('Scatter Plot for each Column (Columns on y-axis)')
plt.show()
```

```python
# plt.figure(figsize=(10,10))
# plt.title('Jewellery Category')
# plt.hist(x=df['Delta'], bins=4000, edgecolor = 'black')
plt.figure(figsize=(10, 7))
plt.title('Jewellery Category')

# Compute histogram
x_values = np.concatenate([df['1D'], df['2D'], df['3D'], df['4D'], df['5D'], df['6D'], df['7D']])

counts, bins, _ = plt.hist(x=x_values, bins=4000, edgecolor='black')
plt.close()  # Close the initial plot to avoid duplicate figures

# Find the non-zero counts for log scale plotting
nonzero_counts = counts[counts > 0]
nonzero_bins = bins[:-1][counts > 0]

# Plot in log-log scale
plt.figure(figsize=(10, 7))
plt.loglog(nonzero_bins, nonzero_counts, linestyle='None', marker='o', markersize=5, color='blue')
# plt.xlim(-5,100)
# plt.ylim(0,50)
plt.xlabel('%change in engagements of Day 7 compared to Day 1')
plt.ylabel('Counts')
```
