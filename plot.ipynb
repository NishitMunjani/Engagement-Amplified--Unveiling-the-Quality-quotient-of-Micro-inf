```python
# Calculate min and max values of the column
min_value = df['Delta'].min()
max_value = df['Delta'].max()

# Apply min-max normalization
df_normalized = df.copy()  # Make a copy to keep the original DataFrame intact
df_normalized['Delta'] = (df['Delta'] - min_value) / (max_value - min_value)
```

```python
df_normalized['Delta'] = df_normalized['Delta']*100
```

```python
df['Engagement'] = df['Engagement'].round(4)
```

```python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

app = Dash(__name__)


app.layout = html.Div([
    html.H4('Interactive scatter plot for Brand Influencer dataset'),
    dcc.Graph(id="scatter-plot"),
    html.P("Filter by Engagement:"),
    dcc.RangeSlider(
        id='range-slider',
        min=0, max=10, step=0.000001,
        marks={0: '0', 10: '10'},
        value=[0, 2],
        tooltip={"placement": "bottom", "always_visible": True}
    ),
])


@app.callback(
    Output("scatter-plot", "figure"), 
    Input("range-slider", "value"))
def update_bar_chart(slider_range):
    low, high = slider_range
    mask = (df_normalized['Engagement'] > low) & (df_normalized['Engagement'] < high)
    fig = px.scatter(
        df_normalized[mask], x="Engagement", y="Days",color="Influencers", size='Delta')
    return fig


app.run_server(debug=True)
```
