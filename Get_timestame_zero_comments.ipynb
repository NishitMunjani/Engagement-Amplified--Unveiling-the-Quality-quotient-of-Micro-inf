```python
# Step 1: Define the directory containing your JSON files
json_dir = 'D://Study//Project//Minor Project//Categories//Clothing//Brand Data'  # Update this path accordingly

# Step 2: List all JSON files in the directory
json_files = [file for file in os.listdir(json_dir) if file.endswith('.json')]

# Step 3: Read each JSON file into a DataFrame and store them in a list
dfs = []
for file_name in json_files:
    file_path = os.path.join(json_dir, file_name)
    df = pd.read_json(file_path)
    dfs.append(df)

# Step 4: Concatenate or merge the DataFrames in the list into one DataFrame
brand_df = pd.concat(dfs, ignore_index=True)
```

```python
def engagement_ratio(likes, followers):
    return likes / followers
```

```python
all_zero_comment_influencers = []

def get_follower_count(inf):
    inf = inf.iloc[0]
    inf = [x.lower() for x in inf]
    for value in inf:
        print("Value:", value)
        # for index, row in microinf_df.iterrows():
        #     for influencer in row['Micro-Influencers']:
        #         print("Influencers:" ,influencer)
        #         if value == influencer:
        #             all_zero_comment_influencers.append(value)
        #             return row['followersCount']
        for index, row in microinf_df.iterrows():
            if value in row['Micro-Influencers']:
                all_zero_comment_influencers.append(value)
                return row['followersCount']
    return None
```

```python
all_engagements = []

for index,row in post_urls_df.iterrows():
    if row['Influencers'] in brand_df['url'].values:
        influencer_row = brand_df[brand_df['url'] == row['Influencers']]
        no_of_likes = influencer_row['likesCount'].iloc[0]
        print("Likes Count:", no_of_likes)
        follower_count = get_follower_count(influencer_row['mentions'])
        print("Followers count:", follower_count)
        engagement = engagement_ratio(no_of_likes, follower_count)
        all_engagements.append(engagement)
    print(index)
```

```python
# Create DataFrame with two columns
df = pd.DataFrame({'Influencers': all_zero_comment_influencers, 'Engagement': all_engagements})
inf_df = df.drop_duplicates()
inf_df_unique = inf_df.groupby('Influencers', as_index=False)['Engagement'].mean()
print(inf_df_unique.shape)
# inf_df_unique.to_csv('averge_zero_comment_influencers_engagement_automobile.csv', index=False)
```
