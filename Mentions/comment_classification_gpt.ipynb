```python
def seconds_remain_to_next_hour():
    now = datetime.now()
    seconds_remaining = 3600 - (now.minute * 60 + now.second + 50)
    return seconds_remaining
```

```python
df = pd.read_excel(file_path)
```

```python
def comment_classification():
    for idx, comment in enumerate(df['comments']):
        if pd.isna(df.iloc[idx]['comment_category']):
            response = get_response(comment)
            if response=='message not available':
                df.loc[idx, 'comment_category'] = None
            if response==None:
                df.to_excel(file_path, index=False)
                break
            df.loc[idx, 'comment_category'] = response
            print(f"{idx}, {comment}, {response}")
            time.sleep(53)
        elif (idx+1)==len(df):
            df.to_excel(file_path, index=False)
            break
```
